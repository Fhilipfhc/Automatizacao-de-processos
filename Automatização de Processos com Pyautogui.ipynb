### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/1mhXZ3JPAnekXP_4vX7Z_sJj35VWqayaR?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# Importando bibliotecas previamente instaladas
import pyautogui as py
import time
import pyperclip as clip
import pandas as pd

py.PAUSE = 1
email=py.prompt(text='Para quem você deseja enviar?', title='Destinatário do Email' , default='exemplo@gmail.com')
py.alert("Não mexa nos componentes do computador durante a execução do código")


#Abrir nova aba
py.hotkey('ctrl','t')
#Digitar o link
drive_link = "https://drive.google.com/drive/folders/1wRTFw0sUVBjRr4hW5U9LF7DjLixRyxym"
caminho=r"C:\Users\juliano\OneDrive\Documentos\python" #Coloque o caminho específico de onde está o seu documento clonado do GitHub
copia = clip.copy(drive_link)
#Colar o link
py.hotkey('ctrl','v')
py.press('enter')
#clicar na tabela
py.click(1075, 427) #Coloque o mouse no documento Vendas - Dez.xlsx e rode a ultima célula
time.sleep(5)
#botão direito
py.rightClick(1075, 427) # Mesma posição do click anterior
time.sleep(3)
#baixar
py.click(917, 617) #Coloque o mouse na posição do botão "Fazer Download" e rode a ultima célula
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\juliano\OneDrive\Documentos\python\Vendas - Dez.xlsx")
display(df)

faturamento = df['Valor Final'].sum()
qtde_produtos = df['Quantidade'].sum()

display(faturamento, qtde_produtos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
py.hotkey('ctrl','t')
#Digitar o link
email_link = "link para acessar seu email - coloque aqui"
clip.copy(email_link)
#Colar o link
py.hotkey('ctrl','v')
py.press('enter')
time.sleep(8)
py.click(96, 171) # Coloque o seu mouse em cima do botão escrever do email e rode a ultima célula
time.sleep(5)
clip.copy(email)
time.sleep(5)
py.hotkey('ctrl','v')
time.sleep(3)
py.press('tab')
time.sleep(3)
assunto = "Faturamentos e quantidade de produtos avaliados"
clip.copy(assunto)
time.sleep(3)
py.hotkey('ctrl','v')
time.sleep(3)
py.press('tab')
time.sleep(3)
mensagem = f"""
    Bom dia, segue os dados de faturamento e quantidade de produtos avaliados.
    O faturamento foi de: R${faturamento}
    A quantidade de produtos avaliados foi de: {qtde_produtos}
"""
clip.copy(mensagem)
time.sleep(3)
py.hotkey('ctrl','v')
time.sleep(3)
py.hotkey('ctrl','enter')
py.hotkey('ctrl','w')
```
